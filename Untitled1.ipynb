```python
from flask import Flask, request, jsonify
from transformers import AutoModel, AutoTokenizer
import torch
import numpy as np

app = Flask(__name__)

# Load pre-trained ProtBERT model
MODEL_NAME = "Rostlab/prot_bert_bfd"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print(f"🔍 Using device: {device}")  # Debug: Check GPU status

# Load model & tokenizer
model = AutoModel.from_pretrained(MODEL_NAME).to(device)
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

@app.route('/generate_vector', methods=['POST'])
def generate_vector():
    try:
        data = request.json
        sequence = data.get("sequence")

        if not sequence:
            return jsonify({"error": "No sequence provided"}), 400

        # Ensure proper tokenization by adding spaces
        formatted_sequence = " ".join(sequence)

        # Tokenize with padding & truncation
        inputs = tokenizer(formatted_sequence, return_tensors="pt", padding=True, truncation=True, max_length=512, return_token_type_ids=False)
        input_ids = inputs["input_ids"].to(device)  # Move to GPU if available
        attention_mask = inputs["attention_mask"].to(device)

        # Check if tokenized sequence is too long
        if input_ids.shape[1] > 512:
            return jsonify({"error": "Sequence is too long, max length is 512 tokens"}), 400

        with torch.no_grad():  # Disable gradients for speed & memory efficiency
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)

        # Extract CLS token embedding (use float16 for efficiency)
        cls_embedding = outputs.last_hidden_state[:, 0, :].squeeze().cpu().numpy().astype(np.float16)

        return jsonify({"embedding": cls_embedding.tolist()})

    except Exception as e:
        return jsonify({"error": str(e)}), 500

if __name__ == '__main__':
    app.run(debug=True, host="0.0.0.0", port=9800)
```
